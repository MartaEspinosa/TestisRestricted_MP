```python
import pandas as pd
import os

annot = pd.read_csv("/users/genomics/marta/TestisProject_SaraRazquin/with_TranscriptomeReconstruction/human/newReference_Resconstructed/1transcript_1gene.reconstructed.csv")
annot
```

```python
TOv3x = pd.read_csv("/projects_eg/projects/marta/TestisRestricted_Microproteins_TSA/with_TranscriptomeReconstruction/Q4_TestisRestricted_TumorSpecific/human/TOv3x_5percent_TestisRestrictedGTEx_Translated_Ctypes.csv")

TAA = pd.read_csv("/projects_eg/projects/marta/TestisRestricted_Microproteins_TSA/with_TranscriptomeReconstruction/Q3_TestisRestrictedTumors/human/TAAdata_long.csv")
TAA = TAA[['transcript_id','gene_id','gene_name','gene_type','ctype']]
TAA.drop_duplicates(inplace=True)

testis = pd.read_csv("/projects_eg/projects/marta/TestisRestricted_Microproteins_TSA/with_TranscriptomeReconstruction/Q2_TestisRestricted/human/testisRestricted_GTEx_translatedONLYtestis.csv")
testis
```

```python
### Ouspenskaia
ouspenskaia = pd.read_csv("/projects_eg/projects/marta/immunopeptidomics_supplementary_data/nuORFdb/Supp8_MHCIpeptides.csv")
## explode columns with separate values
ouspenskaia['Alternative peptides'] = ouspenskaia['Alternative peptides'].str.split(',')
ouspenskaia = ouspenskaia.explode('Alternative peptides')
ouspenskaia['Alternative proteins'] = ouspenskaia['Alternative proteins'].str.split(',')
ouspenskaia = ouspenskaia.explode('Alternative proteins')
## extract transcript_ids
ouspenskaia['transcript_id'] = ouspenskaia['Alternative proteins'].str.split(".", expand=True)[0]
ouspenskaia = ouspenskaia.merge(annot, on="transcript_id")
ouspenskaia ## Alternative peptides
```

```python
### Chong
chong = pd.read_csv("/projects_eg/projects/marta/immunopeptidomics_supplementary_data/Chongetal/Chong_etal_2020_SupData3_41467_2020_14968_MOESM5_ESM.csv", skiprows=1)
chong.rename(columns={'Gene_ID' : 'gene_id'}, inplace=True)
chong = chong.merge(annot, on="gene_id")
chong ## Sequence
```

```python
### IEAtlas
epitopes_IEAtlas = pd.read_csv("/projects_eg/projects/marta/immunopeptidomics_supplementary_data/IEAtlas/epitopes_in_cancer_tissues_IEAtlas.txt", sep="\t")
print(len(epitopes_IEAtlas))
basic_info_IEAtlas = pd.read_csv("/projects_eg/projects/marta/immunopeptidomics_supplementary_data/IEAtlas/epitopes_basic_information_IEAtlas.txt", sep="\t")

IEAtlas = epitopes_IEAtlas.merge(basic_info_IEAtlas, on=["Sequence", "Length", "ORF_ID"])
IEAtlas.rename(columns={'geneId' : 'gene_id'}, inplace=True)
IEAtlas = IEAtlas.merge(annot, on="gene_id")
IEAtlas ## Sequence
```

```python
### Erhard
erhard = pd.read_csv("/projects_eg/projects/marta/immunopeptidomics_supplementary_data/cryptic_peptides_PRISM/S3_cryptic_peptides.csv", skiprows=2)
erhard.rename(columns={'Gene' : 'gene_id'}, inplace=True)

erhard = erhard.merge(annot, on="gene_id")
erhard ## Peptide
```

```python
## de Beijer
Beijer = pd.read_csv("/projects_eg/projects/marta/immunopeptidomics_supplementary_data/deBeijerMTA/tumorassociated_peptides_CTA.csv")
Beijer = Beijer.merge(annot, on=["gene_id","gene_name"])
Beijer ## Peptide
```

```python
## crypticDB
crypticDB = pd.read_csv("/projects_eg/projects/marta/immunopeptidomics_supplementary_data/CrypticProteinDB/immunopeptides_cryptic.csv")
crypticDB['transcript_id'] = crypticDB["transcript.accession"].str.split(".", expand=True)[0]
crypticDB = crypticDB.merge(annot, on="transcript_id")
crypticDB ## Peptide
```

```python
DiazCamarena = pd.read_csv("/projects_eg/projects/marta/immunopeptidomics_TAA_done/uniquePeptides_HCC_CL.FDR5percent.csv")
DiazCamarena['sequence'] = DiazCamarena['sequence'].str.replace(r'\(Oxidation\)', '', regex=True)

DiazCamarena
```

```python
# immuno_combined 
crypticDB_small = crypticDB[['Peptide','gene_id']]
crypticDB_small['source'] = "CrypticDB"

Beijer_small = Beijer[['Peptide','gene_id']]
Beijer_small['source'] = "deBeijer"

erhard_small = erhard[['Peptide','gene_id']]
erhard_small['source'] = "Erhard"

IEAtlas_small = IEAtlas[['Sequence','gene_id']]
IEAtlas_small.columns = ['Peptide','gene_id']
IEAtlas_small['source'] = "IEAtlas"

chong_small = chong[['Sequence','gene_id']]
chong_small.columns = ['Peptide','gene_id']
chong_small['source'] = "Chong"

ouspenskaia_small = ouspenskaia[['Alternative peptides','gene_id']] 
ouspenskaia_small.columns = ['Peptide','gene_id']
ouspenskaia_small['source'] = "nuORFdb"

DiazCamarena_small = DiazCamarena[['sequence','gene_id']]
DiazCamarena_small.columns = ['Peptide','gene_id']
DiazCamarena_small['source'] = "DiazCamarena"

immuno_combined = pd.concat([crypticDB_small, Beijer_small, erhard_small, IEAtlas_small, chong_small, ouspenskaia_small, DiazCamarena_small])
immuno_combined.drop_duplicates(inplace=True)
immuno_combined
```

```python
found = []

# Iterate over each row in df1 (immuno_combined)
for idx, row in immuno_combined.iterrows():
    peptide = row['Peptide']
    
    # Find the ORFpep where the peptide is a substring
    matching_orfpep = TOv3x['ORFpep'].apply(lambda x: peptide in x)
    
    if matching_orfpep.any():
        # Get the first matching ORFpep sequence
        orfpep_seq = TOv3x[matching_orfpep]['ORFpep'].values[0]  # Take the first match
        
        # Add the ORFpep sequence to the row and append to found list
        row_with_orfpep = row.to_dict()  # Convert row to a dictionary
        row_with_orfpep['matching_ORFpep'] = orfpep_seq  # Add the matching ORFpep
        
        found.append(row_with_orfpep)

# Create a new dataframe from the matching rows
peps_found_in_TOv3x = pd.DataFrame(found)

# Display the result
peps_found_in_TOv3x

print(len(set(peps_found_in_TOv3x.matching_ORFpep.values.tolist())), " ORFs contain at least 1 immunoPeptide out of ", len(set(TOv3x.ORFpep.values.tolist())))
```

```python
peps_found_in_TOv3x
```

```python
data_w_immuno = peps_found_in_TOv3x.merge(TOv3x, left_on=['matching_ORFpep'], right_on="ORFpep")
data_w_immuno.drop_duplicates(inplace=True)
data_w_immuno.to_csv("/projects_eg/projects/marta/TestisRestricted_Microproteins_TSA/with_TranscriptomeReconstruction/Q5_immunopeptidomics/human/TOv3x_5percent_TestisRestrictedGTEx_Translated_Ctypes_immuno.csv", index=None)
```

```python
## by gene id
data_immuno_byGene = TOv3x.merge(immuno_combined, on="gene_id")
data_immuno_byGene.drop_duplicates(inplace=True)
data_immuno_byGene.to_csv("/projects_eg/projects/marta/TestisRestricted_Microproteins_TSA/with_TranscriptomeReconstruction/Q5_immunopeptidomics/human/TOv3x_5percent_TestisRestrictedGTEx_Translated_Ctypes_immunobyGene.csv", index=None)
```
