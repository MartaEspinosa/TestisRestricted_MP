```python
## get minimum intron length
blockSizes = bed[[10]]
blockStarts = bed[[11]]

IntronSizesAll = []

for index,row in blockSizes.iterrows():
    ## Split blockSizes and blockStarts for each row
    sepBlockSizes = blockSizes.iloc[index,0].split(',')[:-1]
    sepBlockStarts = blockStarts.iloc[index,0].split(',')[:-1]
    sepBlockEnds = []

    ## Access individual values from sepBlockStarts or sepBlockSizes
    for size, start in zip(sepBlockSizes, sepBlockStarts):
        blockEnd = int(start) + int(size)
        sepBlockEnds.append(blockEnd)  

    sepIntronSize = []
    ## Compute intron size
    if len(sepBlockStarts) > 1:
        for i in range(0,len(sepBlockStarts)-1):
            intronSize = abs(int(sepBlockEnds[i]) - int(sepBlockStarts[i+1]))
            sepIntronSize.append(intronSize)  
            IntronSizesAll.append(intronSize)
    else:
        intronSize = 0
        sepIntronSize.append(intronSize)  

min(IntronSizesAll)
```
