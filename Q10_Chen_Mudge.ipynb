## Chen 2022

```python
Chen2020=pd.read_csv("/data/genomics/marta/Chen_SuppTableS2.csv")
Chen2020
def save_to_fasta(df, filename):
    with open(filename, 'w') as f:
        for index, row in df.iterrows():
            f.write(f">{row['ORF name']}\n")
            f.write(f"{row['Amino acid sequence']}\n")


## select protein coding sequences
Chen2020_to_fasta = Chen2020[['ORF name','Amino acid sequence']]
print(len(Chen2020_to_fasta))

## save the fasta sequences
save_to_fasta(Chen2020_to_fasta, os.path.join(dir,"Q2_TestisRestricted/human/blastp_Chen/Chen_SuppTableS2.fa"))
```

```python
%%bash -s "$dir"

## create blastp database
module load BLAST+/2.12.0-Linux_x86_64

outdir=$1/Q2_TestisRestricted/human/blastp_Chen

makeblastdb -in $outdir/Chen_SuppTableS2.fa -out $outdir/Chen_SuppTableS2 -parse_seqids -dbtype prot
```

```python
%%bash -s "$dir"
## TestisRestricted
module load BLAST+/2.12.0-Linux_x86_64

outdir=$1/Q2_TestisRestricted/human/blastp_Chen

blastp -query $1/Q2_TestisRestricted/human/testisRestricted_GTEx_translatedONLYtestis.NonCanonical.fa \
-db $outdir/Chen_SuppTableS2 \
-task 'blastp-short' \
-out $outdir/TestisRestrictedGTExncORFs_Chen.txt \
-evalue 0.0001 \
-outfmt "6 qseqid sseqid pident length mismatch gapopen qstart qend sstart send evalue bitscore qlen slen qframe sframe sstrand qcovs"
```
